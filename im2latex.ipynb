**The following code is implmentation of the idea from the paper (http://arxiv.org/pdf/1609.04938v1.pdf)
Im2Latex
A general-purpose, deep learning-based system to decompile an image into presentational markup. For example, we can infer the LaTeX or HTML source from a rendered image.**

The basic preprocessing is done similar to the paper.
Used the following code from the repo https://github.com/harvardnlp/im2markup
https://github.com/harvardnlp/im2markup/tree/master/scripts/preprocessing
1)preprocess_formulas.py --> preprocessed formulas  and obtained formulas.norm.lst
2)preprocess_filter.py --> With this script obtained train.lst, test.lst, valid.lst (These files contain the image names and the corresponding formulae indexes
3)preprocess_latex.js --> preprocessed latex formulas
4)generate_latex_vocab.py --> generates required vocab and some important dictionaries files used for further preprocessing

```python
vocab = open('../input/taskset/latex_vocab.txt').read().split('\n')
vocab_to_idx = dict([ (vocab[i],i) for i in range(len(vocab))])
formulas = open('../input/taskset/formulas.norm.lst').read().split('\n')
```

```python
idx_to_vocab = dict([ (i,vocab[i]) for i in range(len(vocab))])
```

```python
def formula_to_indices(formula):
    formula = formula.split(' ')
    res = [0]
    for token in formula:
      if token in vocab_to_idx:
        res.append( vocab_to_idx[token] + 4 )
      else:
        res.append(2)
    res.append(1)
    return res
formulas = list(map( formula_to_indices, formulas))
```

**Building the model**

Building the CNN

```python
model = Sequential()
model.add(Conv2D(16, kernel_size=(3, 3),activation='relu',input_shape=(320,50,1)))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

In the original paper they built the CNN  model implemented from the (Xu et al. 2015) model
We can add multiple layers and can use max pooling layers.
The main idea of implementing a CNN is to learn image features which can be later passed down to LSTM layer

```python
conv_to_rnn_dims = (158,736)
```

**Converting the CNN dimensions to suitable RNN dimensions**

```python
model.add(Reshape(conv_to_rnn_dims))
model.add(Dense(128, activation='relu'))
```

adding the reshape layer is an important step as it involves CNN and LSTM connection
The dense layer used here is the Row Encoder which is similar to the Research paper

```python
model.add(Bidirectional(LSTM(60 , return_sequences = True )))
model.add(Bidirectional(LSTM(60 , return_sequences = True )))
model.add(Dense(503, activation='relu'))
```

The above layers used are the Bi-directional LSTM which are some special type of RNN's.
We can add more LSTM layers which helps in better learning
